```python
import h5py
from PIL import Image
import os

# 配置路径
h5_dir = r'E:\BS\CHIEF\h5'
wsi_dir = r'E:\BS\CHIEF\DATA_DIRECTORY'
save_root = r'E:\BS\CHIEF\patches'
patch_size = 256

# 遍历 h5 文件
for h5_file in os.listdir(h5_dir):
    if not h5_file.endswith('.h5'):
        continue

    patient_id = os.path.splitext(h5_file)[0]
    h5_path = os.path.join(h5_dir, h5_file)
    
    # 寻找对应的 WSI 文件（支持 .tif 或 .svs）
    for ext in ['.tif', '.svs']:
        wsi_path = os.path.join(wsi_dir, patient_id + ext)
        if os.path.exists(wsi_path):
            break
    else:
        print(f"[跳过] 找不到 WSI 文件: {patient_id}")
        continue

    # 保存路径
    save_dir = os.path.join(save_root, patient_id)
    os.makedirs(save_dir, exist_ok=True)

    print(f"处理 {patient_id}...")
    
    try:
        with h5py.File(h5_path, 'r') as f:
            coords = f['coords'][:]

        wsi_img = Image.open(wsi_path)
        w, h = wsi_img.size

        for i, (x, y) in enumerate(coords):
            if x + patch_size > w or y + patch_size > h:
                continue  # 越界 patch 跳过

            patch = wsi_img.crop((x, y, x + patch_size, y + patch_size))
            patch_path = os.path.join(save_dir, f"{patient_id}_patch_{i+1:04d}.png")
            patch.save(patch_path, format='PNG')

            if i % 100 == 0:
                print(f"  已处理 {i+1} 个 patch")

        print(f"[完成] {patient_id} 的 patch 已保存至 {save_dir}\n")

    except Exception as e:
        print(f"[错误] 处理 {patient_id} 时出错: {e}")
```
